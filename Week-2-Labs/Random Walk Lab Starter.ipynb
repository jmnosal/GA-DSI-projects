```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

def walk(R,L):
    walk = np.random.uniform(low = -1, high = 1, size = (R, L)) #first is number of walks: R, second is length L
    walk[:,0] = 0
    return walk.cumsum(axis=1)
```

```python
walk_array = walk(1000,1000)
plt.plot(walk_array.T);
plt.show()
```

```python
plt.hist(walk_array.T[-1]);

```

```python
def DoubleSlit(P, L1, U1, L2, U2):
    walk = np.random.uniform(low = -1, high = 1, size = (1000, 1000))
    walk[:,0] = 0
    walk = walk.cumsum(axis=1)
    
    slit_array = walk[((L1 < walk[:,P]) & (walk[:,P]< U1)) | ((L2 < walk[:,P]) & (walk[:,P]< U2))]
    #Solution code in line below - my original code had an 'or' instead of the '|' and was missing sets of (). 
    #Walk = Walk[((Walk[:, p] > l1) & (Walk[:, p] < u1))| (Walk[:, p] > l2) & (Walk[:, p] < u2)]
    return slit_array

new_array = DoubleSlit( 800, -20, -15, 10, 15)

plt.plot(new_array.T);
plt.show()
```

```python
plt.hist(new_array.T[-1]);
```
