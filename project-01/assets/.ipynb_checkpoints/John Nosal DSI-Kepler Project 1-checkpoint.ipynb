## Step 3: Describe the data

##### 12. Print the min and max of each column

```python
import numpy as np
print "The highest rate is: {0} and the lowest rate is: {1}".format(np.max(rates), np.min(rates))
print "The highest verbal score is: {0} and the lowest verbal score is: {1}".format(np.max(verbal_scores), np.min(verbal_scores))
print "The highest math score is: {0} and the lowest math score is: {1}".format(np.max(math_scores), np.min(math_scores))
```

##### 13. Write a function using only list comprehensions, no loops, to compute Standard Deviation. Print the Standard Deviation of each numeric column.

```python
rate_mean = np.mean(rates)
math_mean = np.mean(math_scores)
verbal_mean = np.mean(verbal_scores)

std_dev_rates = np.sqrt(sum([((x - rate_mean)**2)/len(rates) for x in rates]))
std_dev_math = np.sqrt(sum([((x - math_mean)**2)/len(math_scores) for x in math_scores]))
std_dev_verbal = np.sqrt(sum([((x - verbal_mean)**2)/len(verbal_scores) for x in verbal_scores]))

print std_dev_rates
print std_dev_math
print std_dev_verbal
```
